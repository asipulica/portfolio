clustering using Elbow method for iris dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
iris = datasets.load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])



```

```python
df.head()
df.describe()
df.info()
df.dtypes.value_counts()
df.hist()
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.annotate('Elbow Point,K=5', xy =(5, 50), xytext =(5, 300), arrowprops = dict(facecolor ='green',shrink = 0.05))
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(X_scaled)
```

```python
silhouette_score(X_scaled, kmeans.labels_)
```

## further k-means
costumer segmentation online retail dataset

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

```

```python
df=pd.read_excel('Online-Retail.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
#missed data 
df.isnull().sum()
#to see which columns have missed data
```

```python
#droping rows with missing data
df.dropna(inplace=True)
```

```python
#checking aging
df.info()
```

```python
#RFM marketing analysis to know customer's behavior 
#recency
#frequency
#monetary value
# create customer df and features


#get the total price of each order
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# create customer df and features

customer_df = df.groupby('CustomerID').agg({'InvoiceDate': lambda x: (df['InvoiceDate'].max() - x.max()).days,
                                            'InvoiceNo': 'count',
                                            'TotalPrice': 'sum'})
# rename the features

customer_df.rename(columns={'InvoiceDate': 'recency',
                            'InvoiceNo': 'frequency',
                            'TotalPrice': 'monetary'}, inplace=True)

customer_df.info()
```

```python
#customer df tells us about the customer's purchase
customer_df.sample(10)
```

```python
customer_df.describe()
```

```python
sns.pairplot(customer_df)
```

```python
#from the plot it's obvious the data is not normally distributed we need to normalize it
scaler = MinMaxScaler()
norm_customer = scaler.fit_transform(customer_df)
print(norm_customer)
```

```python
#kmeans
sse = []
#sse:sum of squared error
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=33)
    kmeans.fit(norm_customer)
    sse.append(kmeans.inertia_)
    #elbow thing
plt.plot(range(1, 11), sse, marker='o')
plt.title('Elbow plot')
plt.xlabel('n_clusters')
plt.ylabel('sse')
```

```python
#kmeans model
fin_kmeans = KMeans(n_clusters=3, n_init=10, random_state=33)

fin_kmeans.fit(norm_customer)
```

```python
fin_df = pd.DataFrame(norm_customer, columns=customer_df.columns, index=customer_df.index)

fin_df['cluster'] = fin_kmeans.labels_
fin_df
```

```python
#studying customer behavior and searching for new insights
cluster_mean = fin_df.groupby('cluster')[['recency', 'frequency', 'monetary']].mean().reset_index()

fig, axes = plt.subplots(nrows=3, figsize=(4, 6))

sns.barplot(cluster_mean, x='cluster', y='recency', ax=axes[0])
sns.barplot(cluster_mean, x='cluster', y='frequency', ax=axes[1])
sns.barplot(cluster_mean, x='cluster', y='monetary', ax=axes[2])

plt.tight_layout()
```
